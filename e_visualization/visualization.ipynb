```python
import pandas as pd

# csv 파일 호출
t_df = pd.read_csv('./datasets/titanic.csv')

# 상위 5개 데이터 가져오기
t_df.head(5)
```

```python
# Age 피처의 
t_df[['Age']].describe().T
```

```python
t_df.isna().sum()
```

```python
# 컬럼을 대괄호에 넣지 않고 아래와 같이 표기해도 된다.
t_df.Age = t_df.Age.fillna(t_df.Age.mean())
```

```python
t_df.isna().sum()
```

시각화

```python
t_df.groupby('Pclass')[['Age']].max()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 상세설정 많을 때 bar
# 범주형은 x축
# height는 max 값이다.
plt.bar(x=t_df.Pclass, height=t_df.Age)
plt.show()

# 빠르게 보고싶다면...
sns.barplot(x='Pclass', y='Age', data=t_df)
plt.show()
```

```python
plt.violinplot(t_df.Age)
plt.show()

# hue로 설정할 feature를 x축으로 설정하면, 여러 개의 군짐을 한 번에 볼 수 있다.
sns.violinplot(x='Pclass', y='Age', hue='Pclass', data=t_df)
plt.show()
```

```python
plt.scatter(x=t_df.Age, y=t_df.Fare)
plt.show()

sns.scatterplot(x='Age', y='Fare', hue='Pclass', data=t_df)
plt.show()

# 데이터 프레임으로 접근했을 때 사용..
t_df.plot(x='Age', y='Fare', kind='scatter')
plt.show()
# 여기서 아래처럼 사용하지 않음
# t_df.plot(x='Age', y='Fare', kind='line')
# 선 그래프는> 시간흐름

# 보통은 씨본 많이 사용
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# 막대그래프는 bar플롯이다
# 분포를 나타내는 그래프가 히스토그램

# y축은 count이다. y축 넣을 거 같음 bar 쓰기
# 옆에 나옴
sns.histplot(x='Age', data=t_df, hue='Sex', multiple='dodge')
plt.show()

#stack 쌓는 거
sns.histplot(x='Age', data=t_df, hue='Sex', multiple='stack')
plt.show()

# kde를 통해 그래프가 나옴(분포)
sns.histplot(x='Age', data=t_df, kde=True)
plt.show()
```

```python
plt.hist(t_df.Age)
plt.show()

# rug는 각각의 분포(각각의 밀집도)를 나타낸것
# 중요하지 않음!
sns.distplot(t_df.Age, kde=True, rug=True)
plt.show()

t_df.Age.hist(grid=False)
plt.show()
```

```python
import pandas as pd

# csv 파일 호출
a_df = pd.read_csv('./datasets/avocado.csv')
a_df
```

```python
a_df.isna().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# x축은 수치형(연속형) > 그래서 보통 시간을 많이 사용함
#아래는 날짜형식이 아니라 이상하게 나옴
sns.lineplot(x='Date', y='AveragePrice', data=a_df)
```

```python
a_df.loc[:, 'Date'] = pd.to_datetime(a_df.Date)
line = sns.lineplot(x='Date', y='AveragePrice', data=a_df)
# 날짜 90도로 돌리기
line.set_xticklabels(line.get_xticklabels(), rotation=90)
plt.show()

# 뒤에 있는 잔상..?은 최댓값 최솟값
```

```python
line = sns.lineplot(x='Date', y='AveragePrice', hue='type', data=a_df)
line.set_xticklabels(line.get_xticklabels(), rotation=90)
plt.show()
```
