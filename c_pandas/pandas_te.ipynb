#### DataFrame()
- dict를 DataFrame으로 변환하고자 할 때 DataFrame 생성자에 전달한다.
- 컬럼명을 추가하거나 인덱스명을 변경하는 등 다양하게 설정할 수 있다.

```python
import pandas as pd

film = {
    'title': ['명량', '극한 직업', '범죄 도시3', '국제 시장'],
    'audience': [17_615_919, 16_266_480, 10_682_674, 14_265_222],
    'country': ['한국', '한국', '한국', '한국']
}
film_df = pd.DataFrame(film)
display(film_df)

# 새로운 컬럼명
film_df['income'] = [135_758_658_810, 139_657_105_516, 104_686_489_632, 110_951_970_230]
display(film_df)

# 인덱스명 변경
film_df.index = ['one', 'two', 'three', 'four']
display(film_df)

# 인덱스 초기화
film_df = film_df.reset_index()
display(film_df)

# feature 삭제
film_df = film_df.drop(labels=['index'], axis=1)
display(film_df)

# feature 이름 변경
film_df = film_df.rename(columns={'title': 'name'})
display(film_df)

# 행 삭제
film_df = film_df.drop(index=[2], axis=0)
display(film_df)

# 인덱스 초기화, 기존 인덱스 삭제
film_df.reset_index(drop=True, inplace=True)
display(film_df)
```

#### describe()
- 숫자형 데이터의 개수, 평균, 표준편차, 최소값, 사분위 분포도(중앙값: 50%), 최대값을 제공한다.
- 25번째 백분위수와 75번째 백분위수를 기준으로 정상치의 범위를 설정할 수 있다.

```python
import pandas as pd

happiness_df = pd.read_csv('./datasets/happiness_report_2022.csv')
```

```python
display(happiness_df.describe().T)
```

```python
import numpy as np
from scipy.stats import iqr

happiness_Q1 = np.percentile(happiness_df.score, 25)
happiness_Q3 = np.percentile(happiness_df.score, 75)
print(happiness_Q1, happiness_Q3)

happiness_Q1 = happiness_df.describe().T.loc['score', '25%']
happiness_Q3 = happiness_df.describe().T.loc['score', '75%']
print(happiness_Q1, happiness_Q3)

iqr_value = happiness_Q3 - happiness_Q1
print(iqr_value)

iqr_value = iqr(happiness_df.score)
print(iqr_value)

lower_bound = happiness_Q1 - 1.5 * iqr_value
upper_bound = happiness_Q3 + 1.5 * iqr_value
print(f'정상치 범위: {lower_bound} ~ {upper_bound}')
```

```python
hp_mean_df = happiness_df.groupby('income').mean('score').reset_index()
display(hp_mean_df)

hp_min_df = happiness_df.groupby('income').min('score').reset_index()
display(hp_min_df)

hp_max_df = happiness_df.groupby('income').max('score').reset_index()
display(hp_max_df)

hp_std_df = happiness_df.groupby('income')['score'].std().reset_index()
display(hp_std_df)
```

#### 2022년도 수익에 따른 행복지수 보고서
- "High income" (고소득) 그룹의 경우, 행복점수의 표준편차는 0.552702이다.  
  이는 해당 그룹 내의 행복 점수들이 평균으로부터 약 0.55 만큼 퍼져있다는 것을 의미하고,  
  이 그룹의 사람들은 평균적으로 비교적 비슷한 수준의 행복을 느끼지만 일부 차이가 있을 수 있다.
  
- "Low income" (저소득) 그룹의 경우, 행복점수의 표준편차는 0.723758으로, 고소득 그룹보다 더 큰 변동성을 보인다.  
  이는 이 그룹 내에서 행복 점수들이 더 다양하게 분포되어 있음을 나타낸다.
  
- "Lower middle income" (중하위 소득) 그룹의 표준편차는 0.813456으로, 저소득 그룹보다 더 큰 변동성을 보인다.  
  따라서 이 그룹의 행복 점수들은 더 다양하게 분포되어 있을 것으로 예상된다.
  
- "Upper middle income" (중상위 소득) 그룹의 표준편차는 0.634569로, 중하위 소득 그룹보다는 약간 작지만  
  여전히 상당한 변동성을 보인다. 이는 중상위 소득 그룹 내에서도 행복 점수가 다양하게 분포되어 있음을 시사한다.

📌 요약하자면, 수익 수준이 높을수록 일반적으로 행복 점수의 변동이 적지만, 수익 수준이 낮을수록 그 변동이 더 커진다.

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 4, figsize=(15, 5))

hp_mean_df.plot(title='mean', x='income', y='score', fontsize='4', kind='line', grid=True, ax=axes[0], ylabel="score")
hp_min_df.plot(title='min', x='income', y='score', fontsize='4', kind='line', grid=True, ax=axes[1], ylabel="score")
hp_max_df.plot(title='max', x='income', y='score', fontsize='4', kind='line', grid=True, ax=axes[2], ylabel="score")
hp_std_df.plot(title='std', x='income', y='score', fontsize='4', kind='line', grid=True, ax=axes[3], ylabel="score")
```
